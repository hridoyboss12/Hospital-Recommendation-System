```python
s_list = ["years", "height", "weight", "ap_hi", "ap_lo"]
def standartization(cardio):
    x_std = cardio.copy(deep=True)
    for column in s_list:
        x_std[column] = (x_std[column]-x_std[column].mean())/x_std[column].std()
    return x_std 
x_std=standartization(cardio)
x_std.head()
```

```python
x_melted = pd.melt(frame=x_std, id_vars="cardio", value_vars=s_list, var_name="features", value_name="value", col_level=None)
x_melted
```

```python
plt.figure(figsize=(10,10))
sb.boxplot(x="features", y="value", hue="cardio", data=x_melted)
plt.xticks(rotation=90)
```

```python
ap_list = ["ap_hi", "ap_lo"]
boundary = pd.DataFrame(index=["lower_bound","upper_bound"]) # We created an empty dataframe
for each in ap_list:
    Q1 = cardio[each].quantile(0.25)
    Q3 = cardio[each].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1- 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    boundary[each] = [lower_bound, upper_bound ]
boundary
```

```python
ap_hi_filter = (cardio["ap_hi"] > boundary["ap_hi"][1])
ap_lo_filter = (cardio["ap_lo"] > boundary["ap_lo"][1])                                                           
outlier_filter = (ap_hi_filter | ap_lo_filter)
x_outliers = cardio[outlier_filter]
x_outliers["cardio"].value_counts()
```

```python
sb.countplot(x='cardio',data=x_outliers,linewidth=2,edgecolor=sb.color_palette("dark", 1))
```

```python
out_filter = ((cardio["ap_hi"]>250) | (cardio["ap_lo"]>200) )
print("There is {} outlier".format(cardio[out_filter]["cardio"].count()))
```

```python
cardio = cardio[~out_filter]
```

```python
# splitting dataset into feature data and target data
X = cardio.drop("cardio", axis=1).values
Y = cardio["cardio"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 1)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

mp = MLPClassifier(alpha=1, max_iter=1000)
dec = DecisionTreeClassifier()
ran = RandomForestClassifier(n_estimators=100)
knn = KNeighborsClassifier(n_neighbors=100)
svm = SVC(random_state=1)
naive = GaussianNB()
# gau = GaussianProcessClassifier(1.0 * RBF(1.0))
ada = AdaBoostClassifier()
qu = QuadraticDiscriminantAnalysis()

models = {"Decision tree" : dec,
          "Random forest" : ran,
          "KNN" : knn,
          "SVM" : svm,
          "Naive bayes" : naive,
          "MLPClassifier" : mp,
          "AdaBoostClassifier" : ada,
          "QuadraticDiscriminateAnalysis" : qu}
scores= { }

for key, value in models.items():    
    model = value
    model.fit(X_train, Y_train)
    scores[key] = model.score(X_test, Y_test)
```

```python
scores
```

```python
scores_frame = pd.DataFrame(scores, index=["Accuracy Score"]).T
scores_frame.sort_values(by=["Accuracy Score"], axis=0 ,ascending=False, inplace=True)
scores_frame
```

```python
plt.figure(figsize=(20,5))
sb.barplot(x=scores_frame.index,y=scores_frame["Accuracy Score"])
```
